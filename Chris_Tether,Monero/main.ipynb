# .CSV Import

```python
tether_df = pd.read_csv('../Resources/coin_Tether.csv')
monero_df = pd.read_csv('../Resources/coin_Monero.csv')
```

# Transformation Tether

```python
clean_tether = tether_df.drop(['Name','SNo', 'Volume'], axis = 1)
clean_tether.head()
```

```python
cut_dates = clean_tether['Date'].str.split(" ", n = 1, expand = True)[0]
clean_tether['Date'] = cut_dates
clean_tether.head()
```

```python
cut_yr = clean_tether['Date'].str.split("-", expand = True)
clean_tether['Year'] = cut_yr[0].astype(int)
clean_tether['Bin'] = cut_yr[0] + cut_yr[1]
clean_tether.head()
```

```python
correct_tether = clean_tether.loc[(clean_tether['Year'] >= 2018) &
                                  (clean_tether['Year'] <= 2021)]
correct_tether.head()
```

```python
correct_tether = correct_tether.drop('Year', axis = 1)
correct_tether.head()
```

```python
#calculating the groups
max_tether = correct_tether.groupby('Bin').max()['High']
max_cap_tether = correct_tether.groupby('Bin').max()['Marketcap']
min_tether = correct_tether.groupby('Bin').min()['Low']
avg_open_tether = correct_tether.groupby('Bin').mean()['Open']
avg_close_tether = correct_tether.groupby('Bin').mean()['Close']
```

```python
skele = correct_tether.groupby('Bin').count()
skele['Symbol'] = 'USDT'
skele['High'] = max_tether
skele['Low'] = min_tether
skele['Open'] = avg_open_tether
skele['Close'] = avg_close_tether
skele['Marketcap'] = max_cap_tether
finish_df = skele
finish_df.head()
```

```python
finish_df.drop('Date', axis=1, inplace=True)
```

```python
#dropping to 2021-05 to match datasets
finish_df = finish_df.head(-2)
```

```python
finish_df.reset_index(inplace=True)
finish_df['Name'] = 'Tether'
finish_df.rename(columns = {'Bin': 'Date'})
```

```python
finish_df.to_csv('Output/cleaned_tether.csv', index=False)
```

```python
mean = correct_monero.groupby('Bin').mean()
max = correct_monero.groupby('Bin').max()
min = correct_monero.groupby('Bin').min()
open_monero = mean['Open']
close_monero = mean['Close']
max_monero = max['High']
min_monero = max['Low']
vol_cap_monero = max['Volume']
```

```python
skele = correct_monero.groupby('Bin').count()
skele['Symbol'] = 'XMR'
skele['High'] = max_monero
skele['Low'] = min_monero
skele['Open'] = open_monero
skele['Close'] = close_monero
skele['Marketcap'] = vol_cap_monero * max_monero    
skele.head()
```

```python
skele.drop(['Volume'], axis = 1, inplace=True)
skele.drop(['Date'], axis = 1, inplace=True)
```

```python
skele = skele[['Symbol', 'Open', 'Close', 'High', 'Low', 'Marketcap']]
finished_df = skele.reset_index()
finished_df['Name'] = 'Monero'
finished_df.rename(columns = {'Bin': 'Date'})
```
