```python
doge_df['Date'] = pd.to_datetime(doge_df['Date'])
```

```python
print(type(doge_df['Date'][0]))
```

```python
filtered_doge_df = doge_df.loc[(doge_df['Date'].dt.year >= 2018) & (doge_df['Date'].dt.year <= 2021)]
```

```python
filtered_doge_df.set_index('Date', inplace = True)
```

```python
grouped_filtered_doge_df = filtered_doge_df.groupby([filtered_doge_df.index.year, filtered_doge_df.index.month])
```

```python
grouped_filtered_doge_df.count()
```

```python
final_df_data = []
```

```python
final_df_data.append(grouped_filtered_doge_df['High'].max())

final_df_data.append(grouped_filtered_doge_df['Low'].min())

final_df_data.append(grouped_filtered_doge_df['Open'].mean())

final_df_data.append(grouped_filtered_doge_df['Close'].mean())

final_df_data.append(grouped_filtered_doge_df['High'].max() * grouped_filtered_doge_df['Volume'].max())
```

```python
final_df = pd.DataFrame(final_df_data)
```

```python
final_df = final_df.transpose()
```

```python
final_df.rename(columns={'Unnamed 0': 'Marketcap'}, inplace = True)
```

```python
final_df.index = final_df.index.map('{0[0]}0{0[1]}'.format)
```

```python
final_df.index = final_df.index.astype(int)
```

```python
final_df = final_df.copy().loc[final_df.index <= 202105]
```

```python
final_df['Symbol'] = 'DOGE-USD'
final_df['Name'] = 'Dogecoin'
final_df.reset_index(inplace = True)
final_df.rename(columns={'index': 'Date'}, inplace = True)
final_df
```

```python
final_df.to_csv('coin_Doge_cleaned.csv', index=False)
```
