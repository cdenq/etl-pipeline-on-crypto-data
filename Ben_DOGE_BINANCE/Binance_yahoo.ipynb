```python
#import dependencies
import pandas as pd
import datetime as dt
```

```python
#save file path into variable
csv_file = '../Resources/coin_Binance_yahoo.csv'
```

```python
#read csv
binance_df = pd.read_csv(csv_file)
```

```python
binance_df.drop('Adj Close', axis = 1, inplace = True)
```

```python
binance_df['Date'] = pd.to_datetime(binance_df['Date'])
```

```python
binance_df
```

```python
filtered_binance_df = binance_df.loc[(binance_df['Date'].dt.year >= 2018) & (binance_df['Date'].dt.year <= 2021)]
```

```python
filtered_binance_df.set_index('Date', inplace = True)
```

```python
grouped_filtered_binance_df = filtered_binance_df.groupby([filtered_binance_df.index.year, filtered_binance_df.index.month])
```

```python
grouped_filtered_binance_df.count()
```

```python
final_df_data = []
```

```python
final_df_data.append(round(grouped_filtered_binance_df['High'].max(), 3))

final_df_data.append(round(grouped_filtered_binance_df['Low'].min(), 3))

final_df_data.append(round(grouped_filtered_binance_df['Open'].mean(), 3))

final_df_data.append(round(grouped_filtered_binance_df['Close'].mean(), 3))

final_df_data.append(round(grouped_filtered_binance_df['High'].max() * grouped_filtered_binance_df['Volume'].max(), 3))
```

```python
final_df = pd.DataFrame(final_df_data)
```

```python
final_df = final_df.transpose()
```

```python
final_df.rename(columns={'Unnamed 0': 'Marketcap'}, inplace = True)
```

```python
final_df.index = final_df.index.map('{0[0]}0{0[1]}'.format)
```

```python
final_df.index = final_df.index.astype(int)
```

```python
final_df = final_df.copy().loc[final_df.index <= 202105]
```

```python
final_df['Symbol'] = 'BNB'
final_df['Name'] = 'Binance Coin'
final_df.reset_index(inplace = True)
final_df.rename(columns={'index': 'Date'}, inplace = True)
final_df
```

```python
final_df.to_csv('coin_Binance_yahoo_cleaned.csv', index=False)
```
