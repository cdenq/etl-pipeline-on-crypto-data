```python
#creating a variable from csv file path
eth_df = pd.read_csv('../../Resources/ETH-USD.csv')

#checking dataframe
eth_df.head()
```

Data Cleaning

```python
#checking data types 
eth_df.dtypes

print(type(eth_df['Date'][0]))
```

```python
eth_df['Date']= pd.to_datetime(eth_df['Date'])

#return datetime.date objects without timezone info
eth_df['Date'] = eth_df['Date'].dt.date

eth_df['Date'] = pd.to_datetime(eth_df['Date'])
type(eth_df['Date'][0])
```

```python
#set the index to the Date
eth_df.set_index('Date', inplace = True)
```

```python
eth_df.head()
```

```python
# #filtering for the years 2018 - 2021
eth_year_df =eth_df.loc[(eth_df.index.year >= 2018) & (eth_df.index.year <= 2021)]

eth_year_df.head()
```

```python
group_eth_df = eth_year_df.groupby([eth_year_df.index.year, eth_year_df.index.month])

group_eth_df.count()
```

```python
# #preparing to convert str to int

# eth_df['Date'] = eth_df['Date'].str.replace(r'-', '').astype(int)
# print(eth_df['Date'])
```

```python
clean_eth = []
```

```python
clean_eth.append(group_eth_df['High'].max())
clean_eth.append(group_eth_df['Low'].min())
clean_eth.append(group_eth_df['Open'].mean())
clean_eth.append(group_eth_df['Close'].mean())
clean_eth.append(group_eth_df['High'].max() * group_eth_df['Volume'].max())
```

```python
#Turn it into a Dataframe

clean_eth = pd.DataFrame(clean_eth)
clean_eth = clean_eth.transpose()
```

```python
#Naming the newly created columns
clean_eth.rename(columns={'Unnamed 0': 'Marketcap'}, inplace = True)
```

```python
clean_eth.index = clean_eth.index.map('{0[0]}0{0[1]}'.format)
clean_eth.index = clean_eth.index.astype(int)
```

```python
clean_eth['Symbol'] = 'ETH'
clean_eth['Name'] = 'Ethereum'
clean_eth.reset_index(inplace = True)
clean_eth.rename(columns={'index': 'Date'}, inplace = True)

clean_eth
```
