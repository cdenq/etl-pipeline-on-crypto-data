```python
#converting object to datetime object
bitcoin_df['Date'] = pd.to_datetime(bitcoin_df['Date'])

bitcoin_df['Date'][0].year
```

```python
#dropping Volume as it isn't needed
bitcoin_df.drop(['Volume'], axis=1, inplace= True)
```

```python
#filtering for the years 2018 - 2021
bitcoin_2018_2021_df =bitcoin_df.loc[(bitcoin_df['Date'].dt.year >= 2018) & (bitcoin_df['Date'].dt.year <= 2021)]

```
